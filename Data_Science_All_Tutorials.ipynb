#Set Font Properties for Title and Labels
We can use the fontdict parameter in xlabel(), ylabel(), and title() to set font properties for the title and labels.

```python
import matplotlib.pyplot as plt
import numpy as np
x_point = np.array([10,30,50,70,80,90])
y_point = np.array([20,50,10,70,40,80])
plt.plot(x_point, y_point, color='green')

font1 = {'color':'blue','size':15}
font2 = {'color':'red','size':15}

plt.title('Weather Data in (C*)', fontdict = font1)
plt.xlabel("Kanpur", fontdict = font2)
plt.ylabel("New Delhi",fontdict = font2)
plt.show()
```

#Matplotlib Subplot
Display Multiple Plots
###With the subplot() function we can draw multiple plots in one figure:

The subplot() function takes three arguments that describes the layout of the figure.

The layout is organized in rows and columns, which are represented by the first and second argument.

The third argument represents the index of the current plot.

Ex:

plt.subplot(1, 2, 1)
#Row , Column, Plot No.

#ColorMap
Matplotlib module has a number of available colormaps.

A colormap is like a list of colors, where each color has a value that ranges from 0 to 100.

Here is an example of a colormap:

the colormap with the keyword argument cmap with the value of the colormap, in this case 'viridis' which is one of the built-in colormaps available in Matplotlib.

cmap = 'viridis'


plt.colorbar()

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([10,20,30,40])
y = np.array([20,40,30,50])
colors = np.array([10,20,30,40])
plt.scatter(x,y, cmap = 'viridis', c = colors)

plt.colorbar()
plt.show()
```

#Scatter sizes:

We can change the size of the dots with the s argument.

Just like colors, make sure the array for sizes has the same length as the arrays for the x- and y-axis:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([0,1,13,25,50,60])
y = np.array([20,40,60,80,15,90])
sizes = np.array([100,150,350,400,500,650])
plt.scatter(x,y,sizes)
plt.show()
```

#Alpha
We can also adjust the transparency of the dots with the alpha argument.

Just like colors, make sure the array for sizes has the same length as the arrays for the x- and y-axis:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes, alpha=0.5)

plt.show()
```

#Matplotlib Bars

Creating Bars
With Pyplot, we can use the bar() function to draw bar graphs:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["Apple",'Mango','Cherry','Guava','Banana'])
y = np.array([110,60,140,25,50])

plt.bar(x,y, color= 'hotpink')
plt.title("Fruits Market")
plt.show()
```

#Horizontal Bars use ---barh()
If you want the bars to be displayed horizontally instead of vertically, use the barh() function:

#Explode
The explode parameter, if specified, and not None, must be an array with one value for each wedge.

Each value represents how far from the center each wedge is displayed:


for example i want to explore Kanpur from Pie chart... let's see--below_

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([55, 15, 35, 45])
mylabels = ["kanpur", "Banaras", "New Delhi", "Haryana"]
myExplode = [0.2,0,0,0]
plt.pie(y, labels = mylabels, explode = myExplode)
plt.title('Temperature')
plt.show()
```

#Legend
To add a list of explanation for each wedge, use the legend() function:


```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([55, 15, 35, 45])
mylabels = ["kanpur", "Banaras", "New Delhi", "Haryana"]
myExplode = [0.2,0,0,0]
mycolors = ["Black","hotpink","green","blue"]

plt.pie(y, labels = mylabels, explode = myExplode, shadow =True, colors= mycolors)
plt.title('Temperature Data')
plt.legend()
plt.show()
```

#Legend With Header

To add a header to the legend, add the title parameter to the legend function.

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([55, 15, 35, 45])
mylabels = ["kanpur", "Banaras", "New Delhi", "Haryana"]
myExplode = [0.2,0,0,0]
mycolors = ["Black","hotpink","green","blue"]

plt.pie(y, labels = mylabels, explode = myExplode, shadow =True, colors= mycolors)
plt.legend(title ='city list')
plt.show()
```
