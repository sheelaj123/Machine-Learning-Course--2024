```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
marketing_data = pd.read_csv('bank-additional-full.csv')
```

```python
marketing_data.head(10)
```

```python
marketing_data.info()
```

# 1. Consider the columns - duration, and age as predictors and the column “y”(indicates whether the client will subscribe for a term deposit) as the target variable


```python
x = marketing_data.columns.drop('y')
y = marketing_data['y']
```

```python
from sklearn.preprocessing import MinMaxScaler
#applying MinMaxScaler to the marketing dataset
scaler = MinMaxScaler()
features_to_scale = ["duration", "age"]
scaled_values = scaler.fit_transform(marketing_data[features_to_scale])
marketing_data["duration"] = scaled_values[:,0]
marketing_data["age"] = scaled_values[:,1]
marketing_data.head()
```

#2. Use 60% of the data as training dataset and 40% of data as testing dataset.

```python
from sklearn.model_selection import train_test_split
# selecting the predictors and target
X=marketing_data[['age','duration']]
Y=marketing_data['y']
# splitting data into train and test data set
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.4,random_state=100)
```

3. Build a kNN classification model.

```python
from sklearn.neighbors import KNeighborsClassifier
#kNN with k=3
model = KNeighborsClassifier(n_neighbors=3,metric="euclidean")
```

```python
model.fit(X_train,Y_train)
```

```python
train_accuracy = model.score(X_train,Y_train)
test_accuracy = model.score(X_test,Y_test)
print(train_accuracy,test_accuracy)
#output
#0.921333764972483 0.8775188152464191
```

#Build KNN Model

```python
train_accuracies = []
test_accuracies = []
# building KNN model for values of k from 1 to 100
k_vals = [i for i in range(1,100)]
features = ["age","duration"]
target = "y"
for k in k_vals:
    model = KNeighborsClassifier(n_neighbors=k,metric='euclidean')
    model.fit(X_train,Y_train)
    train_accuracy_k = model.score(X_train,Y_train)
    test_accuracy_k = model.score(X_test,Y_test)
    train_accuracies.append(train_accuracy_k)
    test_accuracies.append(test_accuracy_k)
plt.plot(k_vals,train_accuracies)
plt.plot(k_vals,test_accuracies)
```
